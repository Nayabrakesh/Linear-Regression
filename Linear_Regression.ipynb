```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

#Split  data  into  two  DataFrames  x  and  y  based  on  dependent  and independent variables

```python
#Split  data  into  two  DataFrames  x  and  y  based  on  dependent  and independent variables
x=df[["longitude","latitude","housing_median_age","total_rooms","total_bedrooms","population","households","median_income"]].values
y=df["median_house_value"].values
print("Independent varibles",x)
print("Dependent varibles:",y)
```

#Split x and y into train and test data set  based on test size as 0.2 and  random_state as10

```python
#Split x and y into train and test data set  based on test size as 0.2 and  random_state as10
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
print("x_train and x_test data set shape",x_train.shape,x_test.shape)
print("y_train and y_test data set shape",y_train.shape,y_test.shape)
```

#Print R2 vallue,coefficient and intercept

```python
#fitting the model
regression=LinearRegression()
regression.fit(x_train,y_train)
regression.score(x_test,y_test)
#printing of r-squared values
print("R2 value:",regression.score(x_test,y_test))
print("coefficent:",regression.coef_)
print("intercept:",regression.intercept_)
```

#Compare actual and predicted values.

```python
#comparing of atual values and predicted values
y_pred=regression.predict(x_test)
var=pd.DataFrame({"Actual":y_test,"Predicted":y_pred})
var.head(10)
```
